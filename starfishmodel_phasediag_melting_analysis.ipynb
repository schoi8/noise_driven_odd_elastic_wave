# compare dynamic Lindemann ratio at $\alpha = \alpha_{1}$

```python
base_dir = 'directory for data at alpha=0.79 and vsig=2'

file_name = 'Lratio.txt'
full_path = os.path.join(base_dir,file_name)
with open(full_path,'rb') as f:
    Lratio_vsig2 = np.fromfile(f,np.float64,sep=' ')


base_dir = 'directory for data at alpha=0.79 and vsig=2.4'

file_name = 'Lratio.txt'
full_path = os.path.join(base_dir,file_name)
with open(full_path,'rb') as f:
    Lratio_vsig2p4 = np.fromfile(f,np.float64,sep=' ')

    
base_dir = 'directory for data at alpha=0.79 and vsig=2.5'

file_name = 'Lratio.txt'
full_path = os.path.join(base_dir,file_name)
with open(full_path,'rb') as f:
    Lratio_vsig2p5 = np.fromfile(f,np.float64,sep=' ')
    

base_dir = 'directory for data at alpha=0.79 and vsig=3'

file_name = 'Lratio.txt'
full_path = os.path.join(base_dir,file_name)
with open(full_path,'rb') as f:
    Lratio_vsig3 = np.fromfile(f,np.float64,sep=' ')
    
```

```python
%matplotlib inline

fig,ax = plt.subplots(figsize=(12,6),facecolor=(1,1,1))

ax.plot(np.arange(len(Lratio_vsig2)),Lratio_vsig2,'o-',label='$v_{\sigma}=2.0$')
ax.plot(np.arange(len(Lratio_vsig2p4)),Lratio_vsig2p4,'^-',label='$v_{\sigma}=2.4$')
ax.plot(np.arange(len(Lratio_vsig2p5)),Lratio_vsig2p5,'v-',label='$v_{\sigma}=2.5$')
ax.plot(np.arange(len(Lratio_vsig3)),Lratio_vsig3,'s-',label='$v_{\sigma}=3.0$')
ax.set_xlabel('t',fontsize=20)
ax.set_ylabel('$\gamma_{L}$',fontsize=20)
ax.tick_params(labelsize=15)
ax.legend(fontsize=20)

#plt.savefig('dynamicLindeman.pdf',dpi=200,bbox_inches='tight')
plt.show()
```
